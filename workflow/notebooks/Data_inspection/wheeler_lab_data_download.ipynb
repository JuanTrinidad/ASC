```python
import re

geneid_list = []

with open('../../tmp/whellerlab/text_to_stract_geneID.txt', 'r') as file:
    
    for line in file.readlines():
    
        string = "href=\"protmap.php?id=TvY486_0905250\">cov"

        match = re.search(r'protmap\.php\?id=(.*?)\">cov', line)

        if match:
            #print(match.group(1))
            geneid_list.append(match.group(1))
```

```python
len(geneid_list)
```

```python
from Bio import SeqIO
#funtion to search into fasta file and create cluster fasta files

lenght = []

with open('../../genome_data_sets/query_proteomes/fasta_files/TriTrypDB-63_All_species_clean.fa', "r") as input_file:

    for record in SeqIO.parse(input_file, "fasta"):

        if record.id in geneid_list:
            
            lenght.append([record.id , len(record.seq)] )

            
```
