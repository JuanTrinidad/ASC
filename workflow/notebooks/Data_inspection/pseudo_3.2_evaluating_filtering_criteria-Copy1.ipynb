```python
#tengo que revisar el roden de las columnas, es decir, si por ejemplo panther siempre va de mas inclusivo a menos o como hace.

#modifing columns

def split_get_first(df, col_name, separador, second_separator):
    
    col_name_k = col_name + '_kineto'
    col_name_m = col_name + '_model'
    
    df[col_name_k] = df[col_name_k].str.split(separador,expand=True)[0].str.split(second_separator,expand=True)[0]
    df[col_name_m] = df[col_name_m].str.split(separador,expand=True)[0].str.split(second_separator,expand=True)[0]
    
    return df


def compare_lists(list1, list2):
    
    list1 = list1.split(';')[:-1]
    list2 = list2.split(';')[:-1]

    shared_items = 0
    for item in list1:
        if item in list2:
            shared_items += 1
    
    
    return shared_items
```

# Evaluation of annotation coincidence 

```python
def column_selection_coincidence_calculation(df_rbh_km, columna):
    
    columna_k = columna +'_kineto' #get both columns
    columna_m = columna +'_model'


    df = df_rbh_km[[columna_k, columna_m]].dropna() #droping nana values

    #print(df_rbh_km.shape)
    #print(df.shape)
    
    #calculating exact coincidence
    df['coincidence'] = df[columna_k] == df[columna_m]
    
    #counting coincidence and no coincidence
    data0 = df[df['coincidence'] == True].shape[0]
    data1 = df[df['coincidence'] == False].shape[0]
    
    return data0, data1



```

```python
#new df for simplification
df_looop = df_rbh_km.copy()
#dictionary to storage data
data_colection = {}
#number of iterations as index
iterations = 0


#for exponent in range(0, 22, 2): #creating list of evalues to check
#    evalue = 0.1 ** exponent
#    print(evalue)
    
    #filtering pvalue
for fident in np.arange(0,.75,0.05):    
    fident = round(fident, 3)
    print(fident)
    
    for columna in columns_for_evaluation: #iterating over interesting columns #['PANTHER']:#
        
        for comb in combinations_with_replacement(np.arange(0,1,0.1), 2): #all combination with replacement of coverage values
            
            comb0 = round(comb[0],2)
            comb1 = round(comb[1],2)
            
            #filtering dataframe with parameters of interest
            df_looop = df_looop[(df_looop['fident'] > fident) & (df_looop['COV_query'] > comb0) & (df_looop['COV_target'] > comb1) ]
            
            
            true, false  = column_selection_coincidence_calculation(df_looop, columna)
            
            #colecting results in dictionary
            data_colection[iterations] = [columna, fident, comb0, comb1, true, false]

            #index
            iterations += 1
            
            #print(columna, true, false)
```

# Dataframe generation 

```python
df_all_rbh = pd.DataFrame.from_dict(data_colection, orient='index', columns=['annotation_category', 'fident','COV_query','COV_target','TruePositive','FalsePositive'])

df_only_BH = pd.DataFrame.from_dict(data_colection_BH, orient='index', columns=['annotation_category', 'fident','COV_query','COV_target','TruePositive','FalsePositive'])
```

```python
# precision
def column_calculations(df, name):
    df['sum_TPFP'] = df['TruePositive'] + df['FalsePositive']
    df['precision'] = df['TruePositive'] / df['sum_TPFP']
    df['TruePositive_x_sum_TPFP'] = df['TruePositive'] * df['precision']
    
    df.to_csv(name, sep='\t')
    
    return df
```

```python
df_all_rbh = column_calculations(df_all_rbh, 'rbh_all_hits_annotationDB_evaluation_TP_FP.tsv')
df_only_BH = column_calculations(df_only_BH, 'rbh_only_best_rbh_hit_annotationDB_evaluation_TP_FP.tsv')
```
